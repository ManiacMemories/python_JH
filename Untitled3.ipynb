## 데이터 프레임 정렬

```python
import numpy as np
import pandas as pd
print("DataFrame: ")
data = {'col1' : [2,1,9,8,7,4],
        'col2':['A','A','B',np.nan,'D','C'],
        'col3':[0,1,9,4,2,3]}

index = [1,3,2,5,0,4]
df = pd.DataFrame(data,index=index)
df
```

```python
# index axis 0을 축으로
df = df.sort_index(axis=0)
df
```

```python
df.sort_index(axis=1, ascending=False)
df
```

```python
# sort_value
sorted_df1 = df.sort_values('col1', ascending=True)
sorted_df1
```

```python
# sort_value
sorted_df2 = df.sort_values(['col2','col2'],ascending=True)
sorted_df1
```

## 데이터 분석용 함수

```python
import numpy as np
import pandas as pd
data = {
    'korean' : [50,60,70,30],
    'math' : [30,np.nan,40,20]
}
index = ['a','b','c','d']
df = pd.DataFrame(data,index=index)
print(df.index)
print(df.columns)
df
```

```python
print(df.count(axis=0))
print(df.count(axis=1))
```

```python
print(df.max())
print(df.max(axis=1))
print(df.min())
print(df.min(axis=1))
print(df.sum())
print(df.sum(axis=1))
print(df.mean())
print(df.mean(axis=1))
print(df.mean(axis=1,skipna=False))
```

```python
avg = df['math'].mean()
df['math']=df['math'].fillna(avg)
df
```

## 그룹으로 묶기 (groupby)

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'key' : ['A','B','C','A','B','C'],
    'data1' : [1, 2, 3, 1, 2, 3],
    'data2' : [4, 4, 6, 0, 6, 1],
    # 'data3' : ['안녕', 'a', 'b', '하세요','pple','비']
})

print("DataFrame : ")
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby(['key','data1']).sum() #.reset_index()
```

```python
df.groupby(['key']).aggregate(['min','median','max'])
```

```python
df.groupby(['key']).aggregate({'data1' : 'max', 'data2' : 'sum'})
```
