```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# import matplotlib.date as mdates

import pandas as pd
from datetime import datetime
```

```python
# font 설정 D2coding
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
matplotlib.rcParams["font.family"] = "D2Coding"
```

```python
file_path = "./ref/월별__노선별__차종별_이용차량_통행료수입.csv"
ori = pd.read_csv(file_path, encoding="EUC-KR") # uft-8
ori.columns = ["category","date","cartype1","cartype2"]
ori.head(20)
```

```python
ex1_data = ori.copy()
# ex1_data[ex1_data['category'] == '음식서비스' ]#
obj = ['경부선','남해선','광주대구선','서해안선','울산선','호남선(논산천안선)','중부선(대전통영선)','중부내륙선','영동선','제2영동선','중앙선(대구부산선)','동해선','마산외곽선','경인선',
      '대전남부순환선','중부내륙의지선','광주외곽순환선','중앙선지선','부산외곽순환선','대구외곽순환선'
      ]
ex1_data = ex1_data[ex1_data[ 'category'].isin(obj)]
ex1_data.sample(5)
```

```python
fig , ax = plt.subplots(figsize=(15,10))

category = obj
for cate in category:
    category_data = ex1_data[ex1_data['category'] == cate]
    x_data = category_data['cartype1']
    y_data = category_data['cartype2']
    ax.scatter(x_data,y_data,s = 30, label = cate )

    
ax.set_xlabel("1종")
ax.set_ylabel("2종")
ax.legend(loc="best")
plt.show()
# print(category_data,x_data,y_data)
```
