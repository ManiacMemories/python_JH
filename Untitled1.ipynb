```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#    (D2Coding)
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
matplotlib.rcParams["font.family"] = "D2Coding"

# CSV  
file_path = "./ref/.csv"
coindesk_data = pd.read_csv(file_path, encoding="utf-8")  # utf-8
seq=coindesk_data[['']].to_numpy()
print(' :',len(seq),'\n 5 :',seq[0:5])

seq = coindesk_data[''].to_numpy()

#   
plt.plot(seq,color='blue')
plt.title('Bitcoin Prices (1 year from 2019-02-28)')
plt.xlabel('Days');plt.ylabel('Price in USD')
plt.show()

#      
def seq2dataset(seq,window,horizon):
    X=[]; Y=[]
    for i in range(len(seq)-(window+horizon)+1):
        x=seq[i:(i+window)]
        y=(seq[i+window+horizon-1])
        X.append(x); Y.append(y)
    return np.array(X), np.array(Y)

w=7 #  
h=1 #  

X,Y = seq2dataset(seq,w,h)
print(X.shape,Y.shape)
print(X[0],Y[0]); print(X[-1],Y[-1])
```

```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = "./ref/.csv"
df = pd.read_csv(file_path, encoding="utf-8")

df.columns = ["date", "closing_price", "open", "high", "low", "volume", "change_pct"]
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["days"] = (df["date"] - df["date"].min()).dt.days

# closing_price   (   )
df["closing_price"] = pd.to_numeric(df["closing_price"].str.replace(",", ""), errors="coerce")

#  
plt.figure(figsize=(8, 6))
plt.plot(df["days"], df["closing_price"], color="blue")

# x, y    
plt.xlabel("Days")
plt.ylabel("Price in USD")
plt.title("Bitcoin Prices (1 year from {})".format(df["date"].min().strftime("%Y-%m-%d")))


#  
plt.show()
```
